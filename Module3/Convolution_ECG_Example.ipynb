
## ðŸŽ¯ Filtering an ECG Signal

Many real-world systems â€” from your ears and eyes, to medical devices and electronic circuits â€” process signals over time.

Suppose you're a biomedical engineer working with ECG signals. The raw data you get from a heart monitor contains noise from muscle movement, power lines, and sensors.

We want to extract **clean heartbeats** from this noisy signal. One way to do that is by applying a **low-pass filter** â€” and convolution is how we apply that filter.

Understanding convolution helps you:
- Predict how a system reacts to any input.
- Filter signals to isolate useful features.
- Design and analyze systems in both time and frequency domains.


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Create a noisy ECG-like signal
np.random.seed(0)
t = np.linspace(0, 1, 1000)
clean_ecg = np.sin(2 * np.pi * 5 * t) * np.exp(-5*t)  # simulates a decaying heartbeat
noise = 0.5 * np.random.randn(len(t))
noisy_ecg = clean_ecg + noise

# Simple low-pass filter (moving average kernel)
h = np.ones(20) / 20

# Apply convolution
filtered_ecg = convolve(noisy_ecg, h, mode='same')
```

```python

plt.figure(figsize=(12, 6))
plt.plot(t, noisy_ecg, label='Noisy ECG', alpha=0.5)
plt.plot(t, filtered_ecg, label='Filtered ECG (via Convolution)', linewidth=2)
plt.plot(t, clean_ecg, label='Original Clean ECG', linestyle='--')
plt.title('Using Convolution to Filter ECG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


## âœ… Key Takeaways

- Convolution is **how we apply filters** to signals in the time domain.
- In this example, we used convolution to **smooth a noisy ECG signal**, making it easier to detect heartbeat patterns.
- This same concept is used in heart rate monitors, wearable fitness devices, and medical signal analysis software.

By learning convolution, you're gaining a fundamental tool for analyzing and designing systems that process real-world signals.
